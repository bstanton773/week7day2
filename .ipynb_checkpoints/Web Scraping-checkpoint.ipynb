```python
# Import the requests package and BeautifulSoup class
import requests
from bs4 import BeautifulSoup
```

```python
# Connect to URL you want to connect to
page = requests.get('https://www.nbastuffer.com/2019-2020-nba-player-stats/')
```

```python
page
```

```python
page.text
```

```python
# Instantiate BeautifulSoup class
soup = BeautifulSoup(page.text, 'html.parser')
```

```python
# Retrieve HTML using class
tbody = soup.find(class_='row-hover')
```

```python
# Check type
type(tbody)
```

```python
# Records we want are in each individual tr element
[type(i) for i in tbody]
```

```python
# All data we need is in every odd-numbered indexed row
tr_list = [i for i in tbody if i != '\n']
```

```python
[i for i in tr_list]
```

```python
for column in tr_list[0]:
    print(type(column))
```

```python
data_list = []

# Loop through tr_list
for row in tr_list:
    data_values = []
    # Loop through td element
    for column in row:
        if column != '\n':
        # Append that data to the current record
            data_values.append(column.text)
    # Add the finished row to dataframe
    data_list.append(data_values)
    
data_list
```

```python
class Player:
    _all = []
    
    def __init__(self, name, team, pos, gp, mpg, fta, ftp, tpa, tpp, thpa, thpp, ppg, rpg, apg, spg, bpg, topg):
        self.name = name
        self.team = team
        self.pos = pos
        self.gp = gp
        self.mpg = mpg
        self.fta = fta
        self.ftp = ftp
        self.tpa = tpa
        self.tpp = tpp
        self.thpa = thpa
        self.thpp = thpp
        self.ppg = ppg
        self.rpg = rpg
        self.apg = apg
        self.spg = spg
        self.bpg = bpg
        self.topg = topg
        
        self._all.append(self)
```

```python
for player in data_list:
    Player(
        name=player[1],
        team=player[2],
        pos=player[3],
        gp=player[5],
        mpg=player[6],
        fta=player[10],
        ftp=player[11],
        tpa=player[12],
        tpp=player[13],
        thpa=player[14],
        thpp=player[15],
        ppg=player[18],
        rpg=player[19],
        apg=player[21],
        spg=player[23],
        bpg=player[24],
        topg=player[25]
    )
```
